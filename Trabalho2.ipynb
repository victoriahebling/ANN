```python
# all images are load in variable x...
Arq =['Banana.zip','Orange.zip','Peach.zip','Pitaya.zip',
      'Mango.zip']

img_name=[]
x_gray=[]
x_original = []
for FILE in Arq:
  file_name = zipfile.ZipFile(FILE, 'r')
  file_name.extractall()
  with zipfile.ZipFile(FILE, "r") as f:
      for name in f.namelist():
          img_name.append(name)
```

```python
y_names = []
for yy in img_name:
  name = yy[0]+yy[1]+yy[2]
  y_names.append(name)
```

```python
unique_list = []    
for yy in y_names:
  if yy not in unique_list:
    unique_list.append(yy)
```

```python
print(unique_list)
```

```python
n = len(unique_list)
y = []
for yy in y_names:
  for i in range(n):
    if(yy == unique_list[i]):
      y.append(i)
```

```python
DATASET_SIZE = len(y_names)
Ind = range(DATASET_SIZE)
Img_Size = 80
num_range = DATASET_SIZE
```

```python
Ind_train, Ind_test= train_test_split(Ind,test_size=0.20, shuffle=True, random_state=3)
```

```python
X_train = []
Y_train = []
for II in Ind_train:
  name = img_name[II]
  X_train.append(cv2.resize(cv2.imread(name),(50,50)))
  Y_train.append(y[II])
```

```python
X_test= []
Y_test = []
for II in Ind_test:
  name = img_name[II]
  X_test.append(cv2.resize(cv2.imread(name),(50,50)))
  Y_test.append(y[II])
```

```python
# view some images from the train_ds
num,cols,dim = np.array(X_train[0]).shape
plt.figure(figsize=(10, 10))
for i in range(2):
  ax = plt.subplot(4, 4, i + 1)
  kk = int(random.randrange(num))
  plt.imshow(X_train[kk])
  plt.title(Y_train[kk])
  plt.axis("off")
```
